```python

import pandas as pd
low_memory = False
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
#importando as bibliotecas
```

```python
fileSP = r"C:\analise_gravidez\sampa.xls"
sampa = pd.read_excel(fileSP, usecols=['Nome', 'Data','Valor'])
#lendo os 3 arquivos para comparações
```

```python
sampa.shape
sampa['Data'] = pd.to_datetime(sampa['Data'])

#obtendo como o xls foi referenciado usando sampa como exemplo
```

```python
for par in enumerate(sampa.columns): print (par)
#descrevendo colunas e seus respectivos números
```

```python
maiores = sampa.query('Nome == "Número total de nascidos vivos"') 
menores = sampa.query('Nome == "Número total de nascidos vivos cujas mães tinham 19 anos ou menos"')
#mães maiores e menores de 18 anos agrupadas em suas respectivas variáveis
```

```python
maiores['Data'] = maiores['Data'].dt.strftime('%m/%y')
menores['Data'] = menores['Data'].dt.strftime('%m/%y')
#organizando DateTime
```

```python
menores.groupby('Data')['Valor'].max().tail(15).plot(kind='bar',color = "red", figsize=(12,6))
plt.title('Mães menores de 18 anos em São Paulo')
#gráfico dos nascidos de mães menores de 18 entre os anos 2007-2018
```

```python
maiores.groupby('Data')['Valor'].max().tail(15).plot(kind='bar',color = "blue", figsize=(12,6))
plt.title('Mães maiores de 18 anos em São Paulo')
#gráfico dos nascidos de mães maiores de 18 entre os anos 2007-2018
```

```python
plt.subplots_adjust(wspace=0.2, top=2, right = 3)
plt.subplot(221)
plt.bar(menores.Data,menores.Valor, color='red')
plt.title("Menores")
plt.subplot(222)
plt.bar(maiores.Data,maiores.Valor)
plt.title("Maiores")
#comparativo
```
